```python
# Import Dependencies
import pandas as pd
```

```python
# Read CSV
ff_df = pd.read_csv("../Data/ff_df.csv")
```

```python
# Column to Divide Record Splits
def splitRecord(COLUMN):  
    
    # Empty List to Hold Wins
    win = []
    
    # Empty List to Hold Losses
    loss = []
    
    # Iterate Through Every Row
    for x in range(len(ff_df)):
        
        # Append Wins to Empty List
        win.append(ff_df[COLUMN][x].split("-")[0])
        
        # Append Losses to Empty List
        loss.append(ff_df[COLUMN][x].split("-")[1])
    
    # Return Wins & Losses
    return win, loss
```

```python
# Save Divison Wins
div_win = splitRecord("division_record")[0]

# Save Divison Losses
div_loss = splitRecord("division_record")[1]

# Save Home Wins
home_win = splitRecord("home_record")[0]

# Save Home Losses
home_loss = splitRecord("home_record")[1]

# Save Away Wins
away_win = splitRecord("away_record")[0]

# Save Away Losses
away_loss = splitRecord("away_record")[1]

# Save Divison Wins to DataFrame
ff_df["div_win"] = div_win

# Save Divison Losses to DataFrame
ff_df["div_loss"] = div_loss

# Save Home Wins to DataFrame
ff_df["home_win"] = home_win

# Save Home Losses to DataFrame
ff_df["home_loss"] = home_loss

# Save Away Wins to DataFrame
ff_df["away_win"] = away_win

# Save Away Losses to DataFrame
ff_df["away_loss"] = away_loss

# Save Empty List to Hold Total Wins
total_win = []

# Save Empty List to Hold Total Losses
total_loss = []

# Iterate Through Every Row
for x in range(len(ff_df)):
    
    # Append Total Wins to Empty List
    total_win.append(int(ff_df["home_win"][x]) + int(ff_df["away_win"][x]))
    
    # Append Total Losses to Empty List
    total_loss.append(int(ff_df["home_loss"][x]) + int(ff_df["away_loss"][x]))

# Save Total Wins to DataFrame
ff_df["total_win"] = total_win

# Save Total Losses to DataFrame
ff_df["total_loss"] = total_loss

# Empty List to Hold Record
record = []

# Iterate Through Every Row
for x in range(len(ff_df)):
    
    # Save Win
    wins = ff_df["total_win"][x]
    
    # Save Loss
    losses = ff_df["total_loss"][x]
    
    # Append Record
    record.append(f"{wins}-{losses}")
    
# Save Overall Record to DataFrame
ff_df["overall_record"] = record
```

```python
# Function to Find Times Won While Scoring Under the Week's Average Score
def findLucky(WEEK):
    
        # Save Average Score for the Week
        w1m = WEEK["score"].mean()
        
        # Empty List to Hold Less Than Mean Values
        less_than_mean = []
        
        # Iterate Through Every Row
        for x in range(len(WEEK)):
            
            # Handle Scores Less Than Mean
            if WEEK["score"][x] < w1m:
                
                # Append Name to List
                less_than_mean.append(WEEK["team_name"][x])
                
        # Empty List to Hold Lucky Instances
        lucky = []
        
        # Iterate Through Every Row
        for x in range(len(less_than_mean)):
            
            # Handle When Instance is a Win
            if WEEK[WEEK["team_name"] == less_than_mean[x]]["outcome"].iloc[0] == "Win":
                
                # Append Name to Lucky List
                lucky.append(less_than_mean[x])
                
        # Return Lucky List
        return lucky

# Function to Find Time Lost While Scoring Over the Week's Average Score
def findUnlucky(WEEK):
    
        # Save Average Score for the Week
        w1m = WEEK["score"].mean()
        
        # Empty List to Hold Greater Than Mean Values
        greater_than_mean = []
        
        # Iterate Through Every Row
        for x in range(len(WEEK)):
            
            # Handle Scores Greater Than Mean
            if WEEK["score"][x] > w1m:
                
                # Append Name to List
                greater_than_mean.append(WEEK["team_name"][x])
                
        # Empty List to Hold Unlucky Instances      
        unlucky = []
        
        # Iterate Through Every Row
        for x in range(len(greater_than_mean)):
            
            # Handle When Instance is a Loss
            if WEEK[WEEK["team_name"] == greater_than_mean[x]]["outcome"].iloc[0] == "Loss":
                
                # Append Name to Unlucky List
                unlucky.append(greater_than_mean[x])
                
        # Return Unlucky List  
        return unlucky
```

```python
# Function to Sum Weekly Luck Instances
def completeDF(DF, label):
    
    # Extract Values From Column 1
    c1 = dict(DF[0].value_counts())
    
    # Extract Values From Column 2
    c2 = dict(DF[1].value_counts())
    
    # Extract Values From Column 3
    c3 = dict(DF[2].value_counts())
    
    # Create DataFrame From Extracted Values
    luck_df = pd.DataFrame({"C1":c1, "C2":c2, "C3":c3})
    
    # Sum Extracted Values
    luck_df[f"{label}"] = luck_df.sum(axis=1)
    
    # Sort Values & Convert to DataFrame
    luck_df = luck_df[f"{label}"].sort_values(ascending=False).to_frame()
    
    # Get Full List of Team Names
    full_list = list(ff_df["team_names"].unique())

    # Get Team Names Present in this Luck Instance
    luck_list = list(luck_df.index)

    # Find Team Names Not Present in this Luck Instance
    luck_diff = set(full_list) - set(luck_list)
    
    # Convert Teams Not Present into List
    luck_list = list(luck_diff)
    
    # Empty DataFrame to Hold Zeros
    zeros = []
    
    # Iterate Through Every Team Not Present
    for x in range(len(luck_list)):
        
        # Append Zero to Empty List
        zeros.append(float(0))
        
    # Create DataFrame From List Values 
    luck_df2 = pd.DataFrame(zeros, columns = [f"{label}"], index = luck_list)
    
    # Append Second DataFrame to First
    luck_df = luck_df.append(luck_df2) 
    
    # Return DataFrame
    return luck_df
```
